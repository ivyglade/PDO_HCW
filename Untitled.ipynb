```python
import xarray as xr 
import glob
import pandas as pd
import importlib
import pdo_functions
importlib.reload(pdo_functions)
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
sns.set_palette("colorblind")
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.weight'] = 'light'
rcParams['mathtext.fontset'] = 'cm'
rcParams['mathtext.rm'] = 'serif'
mpl.rcParams["figure.dpi"] = 500
import cartopy.crs as ccrs
import cartopy as ct
import matplotlib.colors as c
import regionmask
import cmasher as cmr
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import matplotlib.patches as mpatches
```

```python
# Open storm reports data 
storm_report_files = sorted(glob.glob('/hurrell-scratch/ivyglade/pdo/storm_reports/StormEvents_details-ftp_v1.0_d*.csv'))
storm_reports = pd.concat([pd.read_csv(file) for file in storm_report_files], ignore_index=True)
```

```python
# Open a single ERA5 file so that we have the coordinates for the desired grid
era5 = xr.open_dataset('/hurrell-scratch/ivyglade/pdo/ERA5_cape/ERA5_cape/1960s/daily_max/e5.oper.an.sfc.128_059_cape.daily_max.19600101-19691231.nc')
```

```python
# Selecting only the conus region from the era5 file
era5_conus = era5['CAPE'].sel(longitude=slice(pdo_functions.convert_longitudes(-125), pdo_functions.convert_longitudes(-66))).sel(latitude=slice(49, 24))
```

```python
# Opening SST data 
sst = xr.open_dataset('/hurrell-scratch/ivyglade/pdo/HadISST_sst.nc')['sst']
```

```python
sst['latitude'].sel(latitude=slice(-5, 5))
```

```python
pdo = pdo_functions.pdo_from_hadisst(sst, 1980, 2010)

nino_34 = pdo_functions.calc_nino_34_timeseries(sst, False, 1980, 2010)
```

```python
# Standardize the Nino3.4 index
nino_34_std = (nino_34 - nino_34.mean()) / nino_34.std()
```

```python
# Convert pdo to xarray for processing
pdo_xr = xr.DataArray(pdo, coords={'time':nino_34_std['time']}, dims='time')

# Select 1980-2024
pdo_1980_2024 = pdo_xr.sel(time=pdo_xr.time.dt.year.isin(np.arange(1980, 2025, 1)))
nino_34_1980_2024 = nino_34_std.sel(time=nino_34_std.time.dt.year.isin(np.arange(1980, 2025, 1)))

# Select only MAM
pdo_1980_2024_mam = pdo_1980_2024.sel(time=pdo_1980_2024.time.dt.month.isin([3, 4, 5]))
nino_34_1980_2024_mam = nino_34_1980_2024.sel(time=nino_34_1980_2024.time.dt.month.isin([3, 4, 5]))
```

```python
# Converting time index to datetime for storm report data
storm_reports['BEGIN_YEARMONTH'] = pd.to_datetime(storm_reports['BEGIN_YEARMONTH'], format='%Y%m')
storm_reports['END_YEARMONTH'] = pd.to_datetime(storm_reports['END_YEARMONTH'], format='%Y%m')
```

```python
# Extract the month from the storm reports data 
storm_reports['MONTH'] = storm_reports['BEGIN_YEARMONTH'].dt.month
```

```python
# specifying by event type
tornados = storm_reports[storm_reports['EVENT_TYPE'].isin(['Tornado'])]
hail = storm_reports[storm_reports['EVENT_TYPE'].isin(['Hail'])]
tstorm_wind = storm_reports[storm_reports['EVENT_TYPE'].isin(['Thunderstorm Wind'])]
```

```python
# Only want MAM
tornados_mam = tornados[tornados['BEGIN_YEARMONTH'].dt.month.isin([3, 4, 5])]
tornados_mar = tornados[tornados['BEGIN_YEARMONTH'].dt.month.isin([3])]
tornados_apr = tornados[tornados['BEGIN_YEARMONTH'].dt.month.isin([4])]
tornados_may = tornados[tornados['BEGIN_YEARMONTH'].dt.month.isin([5])]

hail_mam = hail[hail['BEGIN_YEARMONTH'].dt.month.isin([3, 4, 5])]
hail_mar = hail[hail['BEGIN_YEARMONTH'].dt.month.isin([3])]
hail_apr = hail[hail['BEGIN_YEARMONTH'].dt.month.isin([4])]
hail_may = hail[hail['BEGIN_YEARMONTH'].dt.month.isin([5])]

tstorm_wind_mam = tstorm_wind[tstorm_wind['BEGIN_YEARMONTH'].dt.month.isin([3, 4, 5])]
tstorm_wind_mar = tstorm_wind[tstorm_wind['BEGIN_YEARMONTH'].dt.month.isin([3])]
tstorm_wind_apr = tstorm_wind[tstorm_wind['BEGIN_YEARMONTH'].dt.month.isin([4])]
tstorm_wind_may = tstorm_wind[tstorm_wind['BEGIN_YEARMONTH'].dt.month.isin([5])]
```

```python
# Grid creation for histogram
lat_edges = np.append(era5_conus['latitude'].values-0.25, 23.5)
lon_edges = np.append(era5_conus['longitude'].values-0.25, 294)

# Making a big SE US rectangle
lat_edges_SE = [31, 35]
lon_edges_SE = [pdo_functions.convert_longitudes(-100), pdo_functions.convert_longitudes(-85)]

# Only everyother
lat_edges_0_5 = lat_edges[::2]
lon_edges_0_5 = lon_edges[::2]
```

```python
tornados_mam_count    = np.zeros((135, 50, 118))
tornados_mam_count_SE = np.zeros((135))
for i in range(45):
    mar_torn = tornados_mar.loc[tornados_mar['BEGIN_YEARMONTH'].dt.year == 1980 + i]
    apr_torn = tornados_apr.loc[tornados_apr['BEGIN_YEARMONTH'].dt.year == 1980 + i]
    may_torn = tornados_may.loc[tornados_may['BEGIN_YEARMONTH'].dt.year == 1980 + i]

    tornados_mam_count[i*3]     = pdo_functions.count_in_grid(mar_torn['BEGIN_LAT'], pdo_functions.convert_longitudes(mar_torn['BEGIN_LON']), np.sort(lat_edges_0_5), lon_edges_0_5)
    tornados_mam_count[(i*3)+1] = pdo_functions.count_in_grid(apr_torn['BEGIN_LAT'], pdo_functions.convert_longitudes(apr_torn['BEGIN_LON']), np.sort(lat_edges_0_5), lon_edges_0_5)
    tornados_mam_count[(i*3)+2] = pdo_functions.count_in_grid(may_torn['BEGIN_LAT'], pdo_functions.convert_longitudes(may_torn['BEGIN_LON']), np.sort(lat_edges_0_5), lon_edges_0_5)

    tornados_mam_count_SE[i*3]     = pdo_functions.count_in_grid(mar_torn['BEGIN_LAT'], pdo_functions.convert_longitudes(mar_torn['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
    tornados_mam_count_SE[(i*3)+1] = pdo_functions.count_in_grid(apr_torn['BEGIN_LAT'], pdo_functions.convert_longitudes(apr_torn['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
    tornados_mam_count_SE[(i*3)+2] = pdo_functions.count_in_grid(may_torn['BEGIN_LAT'], pdo_functions.convert_longitudes(may_torn['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
```

```python
hail_mam_count    = np.zeros((135, 50, 118))
hail_mam_count_SE = np.zeros((135))
for i in range(45):
    mar_hail = hail_mar.loc[hail_mar['BEGIN_YEARMONTH'].dt.year == 1980 + i]
    apr_hail = hail_apr.loc[hail_apr['BEGIN_YEARMONTH'].dt.year == 1980 + i]
    may_hail = hail_may.loc[hail_may['BEGIN_YEARMONTH'].dt.year == 1980 + i]

    hail_mam_count_SE[i*3]     = pdo_functions.count_in_grid(mar_hail['BEGIN_LAT'], pdo_functions.convert_longitudes(mar_hail['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
    hail_mam_count_SE[(i*3)+1] = pdo_functions.count_in_grid(apr_hail['BEGIN_LAT'], pdo_functions.convert_longitudes(apr_hail['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
    hail_mam_count_SE[(i*3)+2] = pdo_functions.count_in_grid(may_hail['BEGIN_LAT'], pdo_functions.convert_longitudes(may_hail['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
```

```python
tstorm_wind_mam_count    = np.zeros((135, 50, 118))
tstorm_wind_mam_count_SE = np.zeros((135))
for i in range(45):
    mar_tstorm = tstorm_wind_mar.loc[tstorm_wind_mar['BEGIN_YEARMONTH'].dt.year == 1980 + i]
    apr_tstorm = tstorm_wind_apr.loc[tstorm_wind_apr['BEGIN_YEARMONTH'].dt.year == 1980 + i]
    may_tstorm = tstorm_wind_may.loc[tstorm_wind_may['BEGIN_YEARMONTH'].dt.year == 1980 + i]

    tstorm_wind_mam_count[i*3]     = pdo_functions.count_in_grid(mar_tstorm['BEGIN_LAT'], pdo_functions.convert_longitudes(mar_tstorm['BEGIN_LON']), np.sort(lat_edges_0_5), lon_edges_0_5)
    tstorm_wind_mam_count[(i*3)+1] = pdo_functions.count_in_grid(apr_tstorm['BEGIN_LAT'], pdo_functions.convert_longitudes(apr_tstorm['BEGIN_LON']), np.sort(lat_edges_0_5), lon_edges_0_5)
    tstorm_wind_mam_count[(i*3)+2] = pdo_functions.count_in_grid(may_tstorm['BEGIN_LAT'], pdo_functions.convert_longitudes(may_tstorm['BEGIN_LON']), np.sort(lat_edges_0_5), lon_edges_0_5)

    tstorm_wind_mam_count_SE[i*3]     = pdo_functions.count_in_grid(mar_tstorm['BEGIN_LAT'], pdo_functions.convert_longitudes(mar_tstorm['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
    tstorm_wind_mam_count_SE[(i*3)+1] = pdo_functions.count_in_grid(apr_tstorm['BEGIN_LAT'], pdo_functions.convert_longitudes(apr_tstorm['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
    tstorm_wind_mam_count_SE[(i*3)+2] = pdo_functions.count_in_grid(may_tstorm['BEGIN_LAT'], pdo_functions.convert_longitudes(may_tstorm['BEGIN_LON']), np.sort(lat_edges_SE), lon_edges_SE)[0][0]
```

```python
# Take the mean across the time dimension and then remove 0s
tornados_mam_count_mean    = np.mean(tornados_mam_count, axis=0)
tstorm_wind_mam_count_mean = np.mean(tstorm_wind_mam_count, axis=0)
hail_mam_count_mean        = np.mean(hail_mam_count, axis=0)

tornados_mam_count_mean[tornados_mam_count_mean == 0]       = np.nan
tstorm_wind_mam_count_mean[tstorm_wind_mam_count_mean == 0] = np.nan
hail_mam_count_mean[hail_mam_count_mean == 0]               = np.nan
```

```python
fig, ax = plt.subplots()

ax.plot(tstorm_wind_mam_count_SE)
ax.plot(hail_mam_count_SE)
ax.plot(tornados_mam_count_SE)
```

```python
fig, ax = plt.subplots(1, 3, subplot_kw=dict(projection=ccrs.AlbersEqualArea(central_longitude=-96, central_latitude=37.5)))

tornado_bounds     = [0, 0.02, 0.04, 0.06, 0.08, 0.12, 0.16, 0.2, 0.24, 0.32, 0.4, 0.48, 0.56]
hail_bounds        = [0, 0.04, 0.08, 0.12, 0.16, 0.24, 0.32, 0.4, 0.48, 0.64, 0.8, 0.96, 1.12]
tstorm_wind_bounds = [0, 0.08, 0.16, 0.24, 0.32, 0.48, 0.64, 0.8, 0.96, 1.28, 1.6, 1.92, 2.24]

tornado_norm     = c.BoundaryNorm(tornado_bounds, plt.get_cmap('plasma').N)
hail_norm        = c.BoundaryNorm(hail_bounds, plt.get_cmap('plasma').N)
tstorm_wind_norm = c.BoundaryNorm(tstorm_wind_bounds, plt.get_cmap('plasma').N)

for i in range(3):
    ax[i].coastlines(lw=0.25, color='xkcd:gunmetal')
    ax[i].spines['geo'].set_linewidth(0.25)
    ax[i].spines['geo'].set_edgecolor('xkcd:gunmetal')
    ax[i].add_feature(ct.feature.STATES, lw=0.25, edgecolor='xkcd:gunmetal')

ax[0].pcolormesh(lon_edges_0_5, lat_edges_0_5, np.flipud(tornados_mam_count_mean), transform=ccrs.PlateCarree(), shading='auto', cmap='plasma', norm=tornado_norm)

ax[1].pcolormesh(lon_edges_0_5, lat_edges_0_5, np.flipud(tstorm_wind_mam_count_mean), transform=ccrs.PlateCarree(), shading='auto', cmap='plasma', norm=tstorm_wind_norm)

ax[2].pcolormesh(lon_edges_0_5, lat_edges_0_5, np.flipud(hail_mam_count_mean), transform=ccrs.PlateCarree(), shading='auto', cmap='plasma', norm=tstorm_wind_norm)

cax = plt.axes([0.02, .2, 0.29, 0.02])
cbar = plt.colorbar(mpl.cm.ScalarMappable(cmap='plasma', norm=tornado_norm), cax=cax, orientation='horizontal', spacing='proportional', extend='max', ticks=[0, 0.08, 0.24, 0.4, 0.56])
cbar.set_label(r'tornado reports month$^{-1}$', size=10, fontweight='normal', color='black')
cbar.ax.tick_params(which='both', labelsize=8, width=0.5, length=0, labelcolor='black')
cbar.outline.set_linewidth(0.5)
cbar.outline.set_color('black')
# cbar.outline.set_visible(False)

rect = mpatches.Rectangle(
                         (-100, 31), 
                          -85 - -100, 
                          35 - 31, transform=ccrs.PlateCarree(), fill=False, edgecolor='red')

ax[0].add_patch(rect)

plt.subplots_adjust(left=0,
                    bottom=0.25, 
                    right=1, 
                    top=0.5, 
                    wspace=0.05, 
                    hspace=0.05)
```
