```python
# Coordinates for plotting CONUS404 data
coords = xr.open_dataset('/hurrell-scratch/ivyglade/pdo/conus404/wrf2d_d01_1979-10-12_01:00:00.nc')
```

```python
# Open u and v wind data - can grab monthly data
u_200 = xr.open_mfdataset('/hurrell-scratch/ivyglade/pdo/conus404/u200/daily/conus404_u_200_daily_*.nc', combine='nested', concat_dim='Time')
v_200 = xr.open_mfdataset('/hurrell-scratch/ivyglade/pdo/conus404/v200/daily/conus404_v_200_daily_*.nc', combine='nested', concat_dim='Time')
```

```python
# Calculate mean wind speed
wind_200 = np.sqrt(u_200['__xarray_dataarray_variable__'] **2 + v_200['__xarray_dataarray_variable__']**2)
```

```python
# Select only MAM
wind_200_mam = wind_200.sel(Time=wind_200.Time.dt.month.isin([3, 4, 5]))
```

```python
# Resample to monthly means
wind_200_mam_monthly = wind_200_mam.resample(Time='1M').mean().load()
```

```python
# drop nan months
wind_200_mam_monthly = wind_200_mam_monthly.dropna(dim='Time')
```

```python
# SST data / PDO calculation
sst = xr.open_dataset('/hurrell-scratch/ivyglade/pdo/HadISST_sst.nc')['sst']

pdo = pdo_functions.pdo_from_hadisst(sst, 1980, 2010)

nino_34 = pdo_functions.calc_nino_34_timeseries(sst, False, 1980, 2010)
```

```python
# Nino3.4 standardize
nino_34_std = (nino_34 - nino_34.mean()) / nino_34.std()
```

```python
# Convert pdo to xarray
pdo_xr = xr.DataArray(pdo, coords={'time':nino_34['time']}, dims=['time'])

# Subset only 1980-2022
pdo_1980_2022 = pdo_xr.sel(time=pdo_xr.time.dt.year.isin(np.arange(1980, 2023, 1)))
nino_34_1980_2022 = nino_34_std.sel(time=nino_34_std.time.dt.year.isin(np.arange(1980, 2023, 1)))

# only MAM
pdo_1980_2022_mam = pdo_1980_2022.sel(time=pdo_1980_2022.time.dt.month.isin([3, 4, 5]))
nino_34_1980_2022_mam = nino_34_1980_2022.sel(time=nino_34_1980_2022.time.dt.month.isin([3, 4, 5]))
```

```python
# Align time dimensions of PDO, Nino3.4 and 200 mb winds 
pdo_1980_2022_mam['time'] = nino_34_1980_2022_mam['time']

# wind_200_mam_monthly = wind_200_mam_monthly.rename({'Time':'time'})

wind_200_mam_monthly['time'] = nino_34_1980_2022_mam['time']
```

```python
# Remove the seasonal cycle
wind_200_mam_anoms = wind_200_mam_monthly.groupby('time.month') - wind_200_mam_monthly.groupby('time.month').mean()
```

```python
# Regression
# Grabbing the regression slope
wind_pdo_mam_regress = pdo_functions.linregress_3D(pdo_1980_2022_mam, wind_200_mam_anoms)[2]
wind_nino_mam_regress = pdo_functions.linregress_3D(nino_34_1980_2022_mam, wind_200_mam_anoms)[2]

# Need to grab the correlation coefficient to calculate statistical significance
wind_pdo_mam_cor = pdo_functions.linregress_3D(pdo_1980_2022_mam, wind_200_mam_anoms)[1]
wind_nino_mam_cor = pdo_functions.linregress_3D(nino_34_1980_2022_mam, wind_200_mam_anoms)[1]
```

```python
# Calculate the t statistic
t_stat_pdo = wind_pdo_mam_cor * ((129-2)**0.5) / ((1 - wind_pdo_mam_cor**2)**0.5)
t_stat_nino = wind_nino_mam_cor * ((1292)**0.5) / ((1 - wind_nino_mam_cor**2)**0.5)
```

```python
# Calculate the p-value
p_val_pdo = 2 * (1-scipy.stats.t.cdf(abs(t_stat_pdo), 129-2))
p_val_nino = 2 * (1-scipy.stats.t.cdf(abs(t_stat_pdo), 129-2))
```

```python
# Control for the FDR
adj_p_pdo = pdo_functions.control_FDR(p_val_pdo, 1015, 1367)
adj_p_nino = pdo_functions.control_FDR(p_val_nino, 1015, 1367)
```

```python
pdo_sig = np.where(p_val_pdo < adj_p_pdo, 3, 0)
nino_sig = np.where(p_val_nino < adj_p_nino, 3, 0)
```
